```python
#Counting, Summing, Max, Min

#Counting
count = 0
for staff in ('Head Chef', 'Sou Chef', 'Line cook', 'Cook', 'Helper'):
  count = count + 1
print ('Total staff in kitchen - ', count)

#Sum
Total = 0
for price in (24, 45, 56, 44, 466, 4553):
  Total = Total + price
print ('Total price is ', Total)

#max
max_bp = None
for bp in [120, 130, 110]:
    if max_bp is None or bp > max_bp:
        max_bp = bp
print("Max BP:", max_bp)

#min
RBS = None
for blood_sugar in (78,66,109,210,340,56,71):
    if RBS is None or blood_sugar < RBS :
        RBS = blood_sugar
print ('Hypoglycemia: ', RBS,'mg/dl')
```
